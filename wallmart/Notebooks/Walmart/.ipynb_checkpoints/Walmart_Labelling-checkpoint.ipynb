```python
def read(data_path, sheet_name = ''):
    df = pd.DataFrame()
    if data_path.split('.')[-1] == 'xlsx':
        if sheet_name:
            df = pd.read_excel(data_path, sheet_name=sheet_name)
        else:
            df = pd.read_excel(data_path)
        print("Shape of the data in file {} is {}".format(data_path, df.shape))
    else:
        try:
            df = pd.read_csv(data_path)
            print("Shape of the data in file {} is {}".format(data_path, df.shape))
            if df.shape[0] == 0:
                print("No data in file {}".format(data_path))
        except Exception as e:
            print("Issue while reading data at {} \n{}".format(data_path, e))
    return df


def standardize_date_col(dataframe, date_col):
    dataframe[date_col] = pd.to_datetime(dataframe[date_col], format='%d-%m-%Y', errors='coerce').fillna(pd.to_datetime(df['Date'], format='%d/%m/%y', errors='coerce'))
    # Convert all dates to 'mm-dd-yyyy' format
    dataframe[date_col] = dataframe[date_col].dt.strftime('%Y-%m-%d')
    return dataframe

     
```

## STL for a specific grp 

```python
def aggregate_data(df,state):
    # Aggregating data by date
    data_aggregated = df[df['State']==state].groupby('Date').agg(
        Weekly_Sales=('Weekly_Sales', 'sum'),
        Temperature=('Temperature', 'mean'),
        Fuel_Price=('Fuel_Price', 'mean'),
        CPI=('CPI', 'mean'),
        Unemployment=('Unemployment', 'mean'),
        Holiday_Flag=('Holiday_Flag', 'max') # Assuming if any store has a holiday, it's a holiday across the region
    ).reset_index()
    return data_aggregated


data_aggregated = aggregate_data(df, state)
# Displaying the first few rows of the aggregated data
data_aggregated.head()
```

# STL - For Trend Analysis

"""Seasonal and Trend decomposition using Loess. This is a statistical method of decomposing a Time Series data into 3 components containing seasonality, trend and residual."""

```python
for state in df.State.unique(): print(state)
```

```python
def get_group_data(df, state):
    return df[df['State']==state].groupby(grp_by_cols)["Weekly_Sales"].apply(sum).reset_index()

def prepare_for_stl(group_df):
    data_ = list(group_df['Weekly_Sales'])
    dates = list(group_df['Date'])
    
    data = pd.DataFrame(data_, index=dates)
    data.index = pd.to_datetime(data.index)#,format="%y-%m-%d")
    data.columns = ['Value']
    data = data.asfreq('W-FRI').interpolate(method='linear')
    
    # Padding data to reduce edge effects
    padding_length = 4
    data_padded_data = np.pad(data.Value.tolist(), (padding_length, padding_length), 'edge')
    data_padded_index = np.pad(data.index.tolist(), (padding_length, padding_length), 'edge')
    data_padded = pd.DataFrame(data=data_padded_data, index=data_padded_index)
    return data_padded

def plot_stl(data, trend, seasonal, residual):    
    # Plotting the original data and decomposed components
    plt.figure(figsize=(14, 7))
    plt.subplot(411)
    plt.plot(data, label='Original Data')
    plt.legend(loc='best')
    plt.subplot(412)
    plt.plot(trend, label='Trend')
    plt.legend(loc='best')
    plt.subplot(413)
    plt.plot(seasonal, label='Seasonality')
    plt.legend(loc='best')
    plt.subplot(414)
    plt.plot(residual, label='Residuals')
    plt.legend(loc='best')
    plt.tight_layout()
    plt.show()
    return

def label_group(data_aggregated, trend, seasonal, residual):
    data_aggregated['resid'] = residual.values
    data_aggregated['resid'].fillna(0, inplace=True)
    res_mean = residual.values.mean()
    resid_std = stats.median_abs_deviation(residual.values)
    # print(res_mean,resid_std)
    lower_bound = res_mean - 3*resid_std
    upper_bound = res_mean + 3*resid_std
    
    # Calculate the bounds considering trend and seasonal components if available
    upper_bound_series = upper_bound + trend.values + seasonal.values
    lower_bound_series = lower_bound + trend.values + seasonal.values
    # print(len(upper_bound_series), len(lower_bound_series), upper_bound_series[4],lower_bound_series[4])
    
    # Mark anomalies based on the condition where residue values fall outside the [lower_bound, upper_bound] range
    data_aggregated['Anomaly'] = ((data_aggregated['resid'] < lower_bound) | (data_aggregated['resid'] > upper_bound)).astype(int)    
    # The 'Anomaly' column now has 1 for anomalies and 0 for normal data points
    plot_anomalies(np.array(data_aggregated['Weekly_Sales']), trend, seasonal, residual, lower_bound, upper_bound)
    return data_aggregated

def plot_anomalies(data, trend, seasonal, residual, lower_bound, upper_bound):
    x = residual.index
    y = residual.values    ## residue values from the model - randomness/unpredictable fluctional change from the regular pattern
    
    y[np.isnan(y)] = 0
    ind = np.where([True if i<lower_bound or i>upper_bound else False for i in y])[0]   ## get residue values < -20

    # Main data points
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=x, y=data, mode='lines', name='Data', line=dict(color='black', dash='dash')))
    
    # Anomalies
    for val in ind:
        fig.add_trace(go.Scatter(x=[x[val]], y=[data[val]], mode='markers', name='Anomaly', marker=dict(color='#D63E26', size=10),
                                 text="Anomaly", hoverinfo='text+x+y'))
    
    # Upper and Lower bounds
    upper_bound_series = upper_bound + trend + seasonal
    lower_bound_series = lower_bound + trend + seasonal
    
    fig.add_trace(go.Scatter(x=x, y=upper_bound_series, fill=None, mode='lines', line_color='#A5F9C5', name='Upper Bound'))
    fig.add_trace(go.Scatter(x=x, y=lower_bound_series, fill='tonexty', mode='lines', line_color='#A5F9C5', fillcolor='rgba(190,245,226,0.15)', name='Lower Bound'))
    
    # Update layouts
    fig.update_layout(title="Data Points with Upper and Lower Bounds", xaxis_title="Date", yaxis_title="Value", template="plotly_white", autosize=False, width=1800, height=400)
    
    # Show the plot
    fig.show()

    # # Create a Scatter plot for the residue values
    # fig = go.Figure()
    # fig.add_trace(go.Scatter(x=x, y=y, mode='lines+markers', name='Residues', line=dict(color='black', dash='dash')))
    
    # # Add a vertical line for a specific date
    # # fig.add_vline(x=datetime.strptime("2022-12-23", "%Y-%m-%d").timestamp() * 1000, line_width=3, line_dash="dash", line_color="red")
    
    # # Annotations for anomalies
    # for val in ind:
    #     fig.add_annotation(x=x[val], y=y[val], text="Anomaly", showarrow=True, arrowhead=1, arrowsize=1, arrowwidth=2, arrowcolor="red", ax=40, ay=15)
    
    # # Update layouts
    # fig.update_layout(title="Residue Analysis with Anomalies", xaxis_title="Date", yaxis_title="Residue Value", template="plotly_white", autosize=False, width=1800, height=400)
    
    # # Show the plot
    # fig.show()


def get_labels(df, grp_by_cols):
    labelled_data = pd.DataFrame()
    # Work group wise to get the labels:
    for state in df.State.unique():
        print("\n\n\n", "--"*25,state.upper(),"--"*25)
        group_df = get_group_data(df, state)
        print(group_df.shape)
        # Prepare data:
        prepared_data = prepare_for_stl(group_df)
        result = seasonal_decompose(prepared_data, model='additive',period=7)
        
        # Padding data to reduce edge effects
        padding_length = 4
        trend = result.trend[padding_length:-padding_length]
        seasonal = result.seasonal[padding_length:-padding_length]
        residual = result.resid[padding_length:-padding_length]
        plot_stl(prepared_data, trend, seasonal, residual)
        
        group_df = label_group(aggregate_data(df,state), trend, seasonal, residual)
        # print(group_df.head(3))
        
        labelled_data = pd.concat([labelled_data, group_df])
        print(labelled_data.shape, "Total Labelled Data")
        # print(labelled_data.tail(3))

    print(df.shape, labelled_data.shape, "Original data with store numbers and Final Labelled Data at States level")
    # print(labelled_data.tail(3))

    return labelled_data
        

labelled_data = get_labels(df, grp_by_cols)
```

```python
def clean_and_save_data(labelled_data, filename):
    labelled_data = labelled_data.drop(columns=['resid'],)
    numeric_columns = labelled_data.select_dtypes(include=[int, float]).columns
    # Round the numeric values in these columns to 2 decimal places
    labelled_data[numeric_columns] = labelled_data[numeric_columns].round(2)
    labelled_data.reset_index(drop=True)
    print(labelled_data.head(3))
    # labelled_data.to_csv(filename, index=False)
    return labelled_data
```

```python
labelled_data  = clean_and_save_data(labelled_data, "../Datasets/Walmart_Weekly_labelled.csv")
```
